```python
import pandas as pd
import random
import string
from faker import Faker
from collections import OrderedDict

fake = Faker()
```

```python
random.seed(36)
Faker.seed(36)
```

```python
# Generate Schools data
schools_data = []
list_of_schools = [
    'School of Science and Engineering',
    'John Gokongwei School of Management',
    'School of Humanities',
    'School of Social Sciences',
    'Gokongwei Brothers School of Education and Learning Design'
]

schools_data.append(['school_id', 'school_name'])
for i in range(5):
    schools_data.append([
        i+1, 
        list_of_schools[i]
    ])
```

```python
schools_df = pd.DataFrame(schools_data[1:], columns=schools_data[0])
schools_df
```

```python
# Generate Departments data
departments_data = []
list_of_departments = [
    'Department of Computer Science',
    'Department of Information Systems and Computer Science',
    'Department of Mathematics',
    'Department of Physics',
    'Department of Chemistry',
    'Department of Biology',
    'Department of Psychology',
    'Department of History',
    'Department of Philosophy',
    'Department of English',
    'Department of Filipino',
    'Department of Political Science',
    'Department of Economics',
    'Department of Communication',
    'Department of Legal Management',
    'Department of Marketing',
    'Department of Finance',
    'Department of Entrepreneurship',
    'Department of Educational Leadership and Management',
    'Department of Curriculum, Pedagogy, and Assessment'
]
school_of_dept = [1,1,1,1,1,1,4,4,3,3,3,4,4,4,2,2,2,2,5,5]

departments_data.append(['department_id', 'department_name', 'school_id'])
for i in range(len(list_of_departments)):
    departments_data.append([
        i+1,
        list_of_departments[i],
        school_of_dept[i]
    ])
```

```python
departments_df = pd.DataFrame(departments_data[1:], columns=departments_data[0])
departments_df
```

```python
# Generate Students data
students_data = []
students_data.append([
    'student_id', 
    'student_name', 
    'year_level'
])
for i in range(1, 101):
    students_data.append([
        i, 
        fake.name(),
        random.randint(2, 5)
    ])
```

```python
students_df = pd.DataFrame(students_data[1:], columns=students_data[0])
students_df.head(10)
```

```python
# Generate Classes data
# have to add school_id; class_cost; revenue
classes_data = []
classes_data.append([
    'class_id', 
    'department_id',
    'section_label',
    'rate_per_unit',
    'class_year', 
    'class_semester',
    'class_type'
])

semesters = ['1st', '2nd', 'IS']
years = [2022, 2023]

for i in range(1, 21):
    for j in range(1,7):
        classes_data.append([
            6*i-(6-j), 
            i,
            random.choice(string.ascii_uppercase)+random.choice(string.ascii_uppercase),
            random.choice([4500, 5000]),
            years[(j-1)//3],
            semesters[(j-1)%3], 
            fake.random_element(OrderedDict([("Lecture", 0.9), ("Lab", 0.1)]))
        ])
```

```python
class_df = pd.DataFrame(classes_data[1:], columns=classes_data[0])
class_df.head(10)
```

```python
class_df = class_df.merge(departments_df, on='department_id')
class_df['class_cost'] = pd.Series([
    random.randint(5000,10000) if x == 'Lecture' else random.randint(7500,15000) for x in class_df['class_type']
])
class_df['units'] = pd.Series([
    2 if x == 'Lab' else 3 for x in class_df['class_type']
])
class_df['revenue_per_student'] = class_df['units']*class_df['rate_per_unit']

class_df = class_df[[
    'class_id', 
    'department_id', 
    'school_id', 
    'section_label', 
    'units', 
    'rate_per_unit', 
    'revenue_per_student',
    'class_year', 
    'class_semester',
    'class_type',
    'class_cost'
]]
class_df.head(10)
```

```python
class_df.tail()
```

```python
# Generate Enlistment Fact Table
# assume fully paid, one installment so payment=revenue
enlistment_facts_data = []
enlistment_facts_data.append([
    'student_id',
    'class_id',
    'payment_id'
])

for i in range(1,101):
    sample = random.sample(range(1,121), k=10)
    for j in range(10):
        enlistment_facts_data.append([i, sample[j], 10*(i-1)+(j+1)])
```

```python
enlistment_facts_df = pd.DataFrame(enlistment_facts_data[1:], columns = enlistment_facts_data[0])
enlistment_facts_df.head(20)
```

```python
# Generate Payments data
payments_data = []
payments_data.append([
    'payment_id',
    'payment_status'
])
for i in range(1, len(enlistment_facts_df)+1):
    payments_data.append([
        i, 
        fake.random_element(OrderedDict([("On Time", 0.9), ("Late", 0.1)]))
    ])
```

```python
payments_df = pd.DataFrame(payments_data[1:], columns = payments_data[0])
payments_df.head(10)
```

```python
payments_df = payments_df.merge(enlistment_facts_df, on='payment_id')
payments_df = payments_df[[
    'payment_id',
    'student_id',
    'payment_status'
]]
payments_df.head(10)
```

```python
# Write data to CSV files
schools_df.to_csv('Data/schools.csv', index=False)
departments_df.to_csv('Data/departments.csv', index=False)
students_df.to_csv('Data/students.csv', index=False)
class_df.to_csv('Data/classes.csv', index=False)
payments_df.to_csv('Data/payments.csv', index=False)
enlistment_facts_df.to_csv('Data/enlistment_facts.csv', index=False)
```
