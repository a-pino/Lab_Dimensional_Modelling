```python
import csv
import random
from faker import Faker

fake = Faker()

# Generate Schools data
schools_data = []
list_of_schools = [
    'School of Science and Engineering',
    'John Gokongwei School of Management',
    'School of Humanities',
    'School of Social Sciences',
    'Gokongwei Brothers School of Education and Learning Design'
]

schools_data.append(['school_id', 'school_name'])
for i in range(5):
    schools_data.append([
        i+1, 
        list_of_schools[i]])

# Generate Departments data
departments_data = []
list_of_departments = [
    'Department of Computer Science',
    'Department of Information Systems and Computer Science',
    'Department of Mathematics',
    'Department of Physics',
    'Department of Chemistry',
    'Department of Biology',
    'Department of Psychology',
    'Department of History',
    'Department of Philosophy',
    'Department of English',
    'Department of Filipino',
    'Department of Political Science',
    'Department of Economics',
    'Department of Communication',
    'Department of Legal Management',
    'Department of Marketing',
    'Department of Finance',
    'Department of Entrepreneurship'
]

departments_data.append(['department_id', 'department_name', 'school_id'])
for i in range(18):
    departments_data.append([
        i+1,
        list_of_departments[i],
        random.randint(1, 5)])


# Generate Students data
students_data = []
students_data.append(['student_id', 'student_name', 'enrollment_year'])
for i in range(1, 101):
    students_data.append([
        i, 
        fake.name(), 
        random.randint(2018, 2023)])

# Generate Classes data
classes_data = []
classes_data.append(['class_id', 'department_id', 'school_id', 'class_name', 'class_year', 'class_semester', 'class_cost', 'class_type', 'class_size'])
for i in range(1, 51):
    classes_data.append([
        i, 
        random.randint(1, 10), 
        random.randint(1, 5), 
        fake.random_element([("1st", 0.4), ("2nd", 0.4), ("IS", 0.2)]),
        random.randint(2022, 2023),
        random.randint(1, 5), 
        round(random.uniform(50, 500), 2), 
        fake.random_element([("Lecture", 0.9), ("Lab", 0.1)]),
        random.randint(20, 50)])

# Generate ClassEnrollmentFacts data
class_enrollment_facts_data = []
class_enrollment_facts_data.append([
    'class_id', 
    'class_year', 
    'class_semester', 
    'enrollment_count', 
    'waitlist_count', 
    'withdrawal_count', 
    'tuition_revenue', 
    'fee_revenue', 
    'tuition_increase'])
for i in range(1, 51):
    class_enrollment_facts_data.append([
        i, 
        random.randint(1, 5), 
        random.randint(1, 10), 
        random.randint(1, 100), 
        random.randint(1, 500), 
        random.randint(1, 10), 
        round(random.uniform(5000, 50000), 2), 
        round(random.uniform(1000, 10000), 2), 
        round(random.uniform(0.1, 0.5), 2)])

# Write data to CSV files
def write_to_csv(filename, data):
    with open(filename, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerows(data)

write_to_csv('Data/schools.csv', schools_data)
write_to_csv('Data/departments.csv', departments_data)
write_to_csv('Data/students.csv', students_data)
write_to_csv('Data/classes.csv', classes_data)
write_to_csv('Data/class_enrollment_facts.csv', class_enrollment_facts_data)

print("Data generated and saved to CSV files.")
```
